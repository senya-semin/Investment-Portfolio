```python
import os
import pandas as pd

import plotly.express as px
import matplotlib.pyplot as plt
```

```python
filenames = next(os.walk("Ежегодные данные"), (None, None, []))[2]

trades = pd.DataFrame()
for year in filenames:
    print(year[:4])
    trades = pd.concat([trades, pd.read_excel(f'Ежегодные данные/{year}')], ignore_index=True)
```

```python
trades = trades.iloc[:, 1:]
```

```python
trades.info()
```

```python
trades_type = trades.groupby(by=['Дата', 'Тип']).sum().reset_index()
```

```python
trades_type_short = trades_type.drop(trades_type[pd.DatetimeIndex(trades_type['Дата']).year < 2018].index)
```

```python
df = px.data.gapminder()
fig = px.area(trades_type_short, x="Дата", y="Кол-во сделок", color="Тип", line_group="Тип")
fig.show()
```

```python
df = px.data.gapminder()
fig = px.area(trades_type_short, x="Дата", y="Объем сделок, руб", color="Тип", line_group="Тип")
fig.show()
```

```python
emitent = trades.drop(trades[pd.DatetimeIndex(trades['Дата']).year < 2022].index).pivot_table(values='Цена закрытия', index='Дата', columns='Эмитент')
```

```python
emitent
```

```python
emitent_empty = emitent.isna().sum().sort_values(ascending=False)
```
